# Auxilliary data processing/cleaning functions

```python
def process_age(data, level):
    df = data.copy()
    cols = df.columns
    tokeep = np.array([c for c in cols if 'MOE' not in c and 'Percent' not in c])
    df = df.loc[:, tokeep]
    cols = df.columns
    tokeep = np.array([c for c in cols if 'Male' not in c and 'Female' not in c])
    df = df.loc[:, tokeep]
    cols = df.columns
    tokeep = np.array([c for c in cols if 'SELECTED AGE CATEGORIES' not in c])
    df = df.loc[:, tokeep]
    cols = df.columns
    tokeep = np.array([c for c in cols if 'PERCENT ALLOCATED' not in c])
    df = df.loc[:, tokeep]
    cols = df.columns
    tokeep = np.array([c for c in cols if 'SUMMARY INDICATORS' not in c])
    df = df.loc[:, tokeep]
    cols = df.columns
    df.columns = [c.split('!!')[-1] for c in cols]
    cols = df.columns
    df.columns = [c.replace('years', '') for c in cols]
    df = df.rename(columns={df.columns[1]:'name', df.columns[2]:'total'})
    df = getrows(df, level)
    return df
```

```python
def process_income(data, level):
    df = data.copy()
    cols = df.columns
    tokeep = np.array([c for c in cols if 'Margin of Error' not in c])
    df = df.loc[:, tokeep]
    cols = df.columns
    df.columns = [c.split('!!')[-1] for c in cols]
    df = df.rename(columns={df.columns[1]:'name', df.columns[2]:'total'})
    df = getrows(df, level)
    return df
```

```python
def process_industry(data, level):
    df = data.copy()
    cols = df.columns
    tokeep = np.array([c for c in cols if 'Margin of Error' not in c])
    df = df.loc[:, tokeep]
    cols = df.columns
    tokeep = np.array([c for c in cols if 'Male' not in c and 'Female' not in c])
    df = df.loc[:, tokeep]
    cols = df.columns
    df.columns = [c.split('!!')[-1] for c in cols]
    df = df.rename(columns={df.columns[1]:'name', df.columns[2]:'total'})
    df = getrows(df, level)
    return df
```

```python
def process_occupation(data, level):
    df = data.copy()
    cols = df.columns
    tokeep = np.array([c for c in cols if 'Margin of Error' not in c])
    df = df.loc[:, tokeep]
    cols = df.columns
    tokeep = np.array([c for c in cols if 'Male' not in c and 'Female' not in c])
    df = df.loc[:, tokeep]
    cols = df.columns
    df.columns = [c.split('!!')[-1] for c in cols]
    df = df.rename(columns={df.columns[1]:'name', df.columns[2]:'total'})
    df = getrows(df, level)
    return df
```

```python
def process_commute(data, level):
    df = commute_city_raw.copy()
    cols = df.columns
    tokeep  = [c for c in cols if 'moe' not in c.lower()]
    tokeep = [c for c in tokeep if 'Male' not in c and 'Female' not in c]
    tokeep = [c for c in tokeep if 'TIME LEAVING HOME TO GO TO WORK' not in c]
    tokeep = [c for c in tokeep if 'percent allocated' not in c.lower()]
    tokeep = [c for c in tokeep if 'vehicles available' not in c.lower()]
    df = df.loc[:, tokeep]
    df = df.rename(columns={df.columns[1]:'name', df.columns[2]:'total'})
    df = df.rename(columns={'Estimate!!Total!!Workers 16 years and over who did not work at home':'total_notwfh'})
    df.columns = [c.replace('Estimate!!Total!!Workers 16 years and over!!', '') for c in df.columns]
    df.columns = [c.replace('Estimate!!Total!!Workers 16 years and over who did not work at home!!','') for c in df.columns]
    df = getrows(df, level)
    return df
```

```python
def getrows(data, level):
    df = data.copy()
    if level == 0:
        cleaner = lambda n: n.lower().replace('county, california', '').strip()
        df['name'] = df['name'].apply(cleaner)
        tokeep = [n in baycounties for n in df['name']]
        df = df.loc[tokeep,:]
        df = df.sort_values('name').reset_index(drop=True)
        df = df.rename(columns={'name':'county'})
    elif level == 1:
        cleaner = lambda n: n.lower().replace(', california', '').strip()
        df['name'] = df['name'].apply(cleaner)
        tokeep = [n in eastalameda for n in df['name']]
        df = df.loc[tokeep,:]
        df = df.sort_values('name').reset_index(drop=True)
        df = df.rename(columns={'name':'city'})
    elif level == 2:
        cleaner = lambda n: n.lower().replace(', san francisco county, california','').replace('census tract','').strip()
        df['name'] = df['name'].apply(cleaner)
        df = df.sort_values('name').reset_index(drop=True)
        df = df.rename(columns={'name':'tract'})
    elif level == 42:
        cleaner = lambda n: n.lower().replace(', california', '').strip()
        df['name'] = df['name'].apply(cleaner)
        df = df.sort_values('name').reset_index(drop=True)
        df = df.rename(columns={'name':'city'})
    df = df.drop(columns='id')
    return df
```

# Process ridership data

```python
entry_feb = np.array(riders_feb_raw.loc['Entries',:])[:50].reshape(-1,1)
exit_feb = np.array(riders_feb_raw.loc[:,'Exits'])[:50].reshape(-1,1)
entry_mar = np.array(riders_mar_raw.loc['Entries',:])[:50].reshape(-1,1)
exit_mar = np.array(riders_mar_raw.loc[:,'Exits'])[:50].reshape(-1,1)
entry_apr = np.array(riders_apr_raw.loc['Entries',:])[:50].reshape(-1,1)
exit_apr = np.array(riders_apr_raw.loc[:,'Exits'])[:50].reshape(-1,1)
ridership = np.hstack((entry_feb, exit_feb, entry_mar, exit_mar, entry_apr, exit_apr))
ridership.shape
```

```python
cols = ['entry_feb', 'exit_feb', 'entry_mar', 'exit_mar', 'entry_apr', 'exit_apr']
ridership_df = pd.DataFrame(data=ridership, index=riders_feb_raw.index[:50], columns=cols)
ridership_df['entrydelta_fm'] = np.abs(ridership_df['entry_mar'] - ridership_df['entry_feb'])
ridership_df['entrydelta_ma'] = np.abs(ridership_df['entry_apr'] - ridership_df['entry_mar'])
ridership_df['entrydelta_fa'] = np.abs(ridership_df['entry_apr'] - ridership_df['entry_feb'])
ridership_df['exitdelta_fm'] = np.abs(ridership_df['exit_mar'] - ridership_df['exit_feb'])
ridership_df['exitdelta_ma'] = np.abs(ridership_df['exit_apr'] - ridership_df['exit_mar'])
ridership_df['exitdelta_fa'] = np.abs(ridership_df['exit_apr'] - ridership_df['exit_feb'])
ridership_df.shape
```

```python
ridership_df.to_excel('processed/ridership.xlsx')
```

```python
dropcols = ['FY', 'Qtr', 'Month', 'Day', 'Date', 'Unnamed: 56', 'Total']
```

```python
riders_tot = riders_tot_raw[riders_tot_raw['FY'] == 'FY20']
riders_tot.head(3)
```

```python
def tot_riders(month):
    riderdf = riders_tot[riders_tot['Month']==month]
    riderdf = riderdf.drop(columns=dropcols)
    ridermat = np.array(riderdf)
    ridertot = ridermat.sum(axis=0).reshape(-1,1)
    return ridertot
```

```python
riders_feb_tot = tot_riders('February')
riders_mar_tot = tot_riders('March')
riders_apr_tot = tot_riders('April')
delta_fm = np.abs(riders_mar_tot - riders_feb_tot)
delta_ma = np.abs(riders_apr_tot - riders_mar_tot)
delta_fa = np.abs(riders_apr_tot - riders_feb_tot)
riders = np.hstack((riders_feb_tot, riders_mar_tot, riders_apr_tot, delta_fm, delta_ma, delta_fa))
riders.shape
```

```python
riders_tot.drop(columns=dropcols).columns
```

```python
idx = riders_tot.drop(columns=dropcols).columns
cols = ['feb', 'mar', 'apr', 'delta_fm', 'delta_ma', 'delta_fa']
ridership_df = pd.DataFrame(data=riders, index=idx, columns=cols)
ridership_df.head(3)
```

```python
ridership_df.to_excel('processed/ridership_tot.xlsx')
```
