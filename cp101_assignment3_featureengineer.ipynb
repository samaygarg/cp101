# Generate features

```python
def get_totals(df):
    total = np.array(df.iloc[:,1])
    total = np.where(total!=0, total, np.ones(len(total)))
    total = total.reshape(-1,1)
    return total

def get_counts(df):
    return np.array(df.iloc[:,2:])

def get_weights(df):
    counts = get_counts(df)
    totals = get_totals(df)
    return counts/totals
```

## Age features

```python
age_raw.iloc[:,2:].head(3)
```

```python
age_weights = get_weights(age_raw)
age_weights.shape
```

```python
age_mids = np.arange(2.5, 87.7, 5)
age_mids.shape
```

```python
age_mean = np.sum((age_weights*age_mids), axis=1).reshape(-1,1)
age_mean.shape
```

```python
age_u20 = np.sum(age_weights[:,:4], axis=1).reshape(-1,1)
age_u30 = np.sum(age_weights[:,4:6], axis=1).reshape(-1,1)
age_u50 = np.sum(age_weights[:,6:10], axis=1).reshape(-1,1)
age_u65 = np.sum(age_weights[:,10:13], axis=1).reshape(-1,1)
age_u99 = np.sum(age_weights[:,13:], axis=1).reshape(-1,1)
```

```python
age_cat = np.hstack((age_mean, age_u20, age_u30, age_u50, age_u65, age_u99))
age_cat.shape
```

```python
age_mean.shape
```

```python
age_all = age_weights
age_all.shape
```

## Race features

```python
race_raw.iloc[:,2:].head(3)
```

```python
race_mean = np.hstack((race_white, race_black, race_asian, race_other))
race_mean.shape
```

# Convert feature matrix to station-basis

```python
badstations = ['SO', 'OA', 'WS', 'ML', 'BE', 'PC', 'AN']
ridership_df = ridership_raw.drop(badstations)
homes_df = homes_raw.drop(badstations)
ridership_df.shape, homes_df.shape
```

```python
homes_norm = np.array(homes_df) / np.array(homes_df['total']).reshape(-1,1)
homes_df = pd.DataFrame(data=homes_norm, index=homes_df.index, columns=homes_df.columns)
homes_df = homes_df.fillna(0)
homes_df = homes_df.drop(columns='total')
homes_df.head(3)
```

```python
baycities = list(homes_df.columns)
baycities = [x.lower() for x in baycities]
len(baycities)
```

```python
def get_rows(df):
    tokeep = [False for c in df['city']]
    i = 0
    for c in df['city']:
        for n in baycities:
            if n in c:
                tokeep[i] = True
        i += 1
    todrop = ['el sobrante cdp (riverside county)', 'west menlo park cdp', 'west modesto cdp',
              'west sacramento city', 'mountain view acres cdp', 'mountain view cdp', 'south san jose hills cdp',
              'university of california-davis cdp', 'east palo alto city', 'east richmond heights cdp',
              'el cerrito cdp', 'lake davis cdp', 'franklin cdp (sacramento county)',
              'north richmond cdp', 'los alamos cdp']
    df = df.loc[tokeep,:]
    df = df.set_index('city')
    df = df.drop(todrop)
    return df
```

```python
bay_mean_df = get_rows(feature_mean_df)
bay_cat_df = get_rows(feature_cat_df)
bay_all_df = get_rows(feature_all_df)
bay_mean_df.shape, bay_cat_df.shape, bay_all_df.shape
```

```python
homes_df.shape
```

```python
bay_mean = np.array(bay_mean_df)
bay_cat = np.array(bay_cat_df)
bay_all = np.array(bay_all_df)
homes = np.array(homes_df)
bay_mean_df.shape, bay_cat_df.shape, bay_all_df.shape, homes_df.shape
```

```python
station_mean = homes @ bay_mean
station_cat = homes @ bay_cat
station_all = homes @ bay_all
station_mean.shape, station_cat.shape, station_all.shape
```

```python
station_mean_df = pd.DataFrame(data=station_mean, columns=bay_mean_df.columns, index=homes_df.index)
station_cat_df = pd.DataFrame(data=station_cat, columns=bay_cat_df.columns, index=homes_df.index)
station_all_df = pd.DataFrame(data=station_all, columns=bay_all_df.columns, index=homes_df.index)
station_mean_df.head(3)
```
